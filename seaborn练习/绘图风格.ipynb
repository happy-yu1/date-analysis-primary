### set_context( )--大小
context=None, font_scale=1, rc=None

```python
exercise=sns.load_dataset('exercise')
exercise.plot()
sns.set_context('notebook') 
```

```python
exercise=sns.load_dataset('exercise')
exercise.plot()
sns.set_context('poster') 
```

```python
exercise=sns.load_dataset('exercise')
exercise.plot()
sns.set_context('paper')  #  线稍微细一点
```

```python
exercise=sns.load_dataset('exercise')
exercise.plot()
sns.set_context('talk') 
```

#### font_scale参数-设置字体大小，

```python
exercise=sns.load_dataset('exercise')
exercise.plot()
sns.set_context("talk",font_scale=2)
```

#### rc={'line.linewidth':2}-设置线的粗细
传入的是一个字典

```python
exercise=sns.load_dataset('exercise')
exercise.plot()
sns.set_context("talk", rc={"lines.linewidth": 5})
```

## despine()
通过参数设置，只展示部分样式，true为不显示
fig=None, ax=None, top=True, right=True, left=False, bottom=False, offset=None, trim=False,

```python
exercise=sns.load_dataset('exercise')
sns.boxplot(x='diet',y='pulse',data=exercise)
sns.set(style='ticks')      #法二
```

### 默认情况

```python
exercise=sns.load_dataset('exercise')
sns.boxplot(x='diet',y='pulse',data=exercise)
sns.set(style='ticks')      #法二
sns.despine()  #将右边和上边的线去掉了
```

```python
exercise=sns.load_dataset('exercise')
sns.boxplot(x='diet',y='pulse',data=exercise,palette='pastel')
#palette用来调整图表的整体颜色，有 deep,muted, pastel, bright, dark, colorblind
sns.despine(left=True,bottom=True)  #将右.左，上，下边的线去掉了
```

### offset参数
设置图形离刻度线的距离

```python
sns.violinplot(x='diet',y='pulse',data=exercise)
```

```python
sns.violinplot(x='diet',y='pulse',data=exercise)
sns.despine(offset=20)
```

## with语句-子图样式不同
with语句内可以设置不同的样式，语句外为其他样式

```python
import tushare as ts
pro=ts.pro_api('5f2dc9d17ebed3601a4fbea74148a25067d067138bf88515693e9b7d')
quotes=pro.daily(ts_code='600015.SH', start_date='20180101', end_date='20190101',fields=('trade_date','open','close','high','low'))
quotes['trade_date']=pd.to_datetime(quotes['trade_date'])
quotes.set_index('trade_date',inplace=True)     #将日期变成列索引，有利于后期x轴刻度的调整
quotes.sort_index(ascending=True,inplace=True)     #正序排序
quotes.head(2)
```

```python
plt.figure(figsize=(10,4))
with sns.axes_style('whitegrid'):
    plt.subplot(2,2,1)
    sns.boxplot(data=quotes)
plt.subplot(2,2,2)
sns.boxplot(data=quotes)
plt.tight_layout(2)
```
