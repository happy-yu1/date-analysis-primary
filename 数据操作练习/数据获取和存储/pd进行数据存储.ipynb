```python
import tushare as ts
import pandas as pd
pro=ts.pro_api('5f2dc9d17ebed3601a4fbea74148a25067d067138bf88515693e9b7d')
```

### 一、数据存入

### 1、从tushare中传入数据

```python
df= pro.daily(ts_code='000001.SZ', start_date='20180701', end_date='20180718')
df.head()
```

### 2. pd存储为csv类型

```python
df.to_csv('000001.SZ.csv')
```

### 3.存储为json

```python
df.to_json('000001.SZ.json')
```

### 4.存储为excel类型

```python
df.to_excel('000001.SZ.xls')
```

### 5.存储为txt格式

```python
df.to_csv('000001.SZ.txt')
```
