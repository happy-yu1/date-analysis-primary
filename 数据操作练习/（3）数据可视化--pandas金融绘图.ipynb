```python
import numpy as np
import pandas as pd
import tushare as ts
pro=ts.pro_api('5f2dc9d17ebed3601a4fbea74148a25067d067138bf88515693e9b7d')
df=pro.daily(ts_code='600015.SH', start_date='20180101', end_date='20190101',fields=('trade_date','open','close','high','low'))
df['trade_date']=pd.to_datetime(df['trade_date'])
df.set_index('trade_date',inplace=True)  
df.head()
```

## 情形一

```python
df.plot( subplots=True,sharex=True, sharey=True,layout=(2,2), figsize=(15,3),title='股票行情数据' ,use_index=True,legend=True, fontsize=13)
```

```python
df.plot(kind='kde', subplots=True,sharex=True, sharey=True,layout=(2,2), figsize=(15,3),title='股票行情数据' ,use_index=True,legend=True, fontsize=13)
```

## 情形二-添加ax

```python
ax1=df.plot(kind='scatter', x='close',y='high',figsize=(15,3),label='Class1' ,legend=True, fontsize=13)
df.plot(kind='scatter', x='close',y='low',figsize=(15,3),ax=ax1,label='Class2' ,color='r',legend=True, fontsize=13)
```

# 随机数

```python
import numpy.random as npr
```

## 标准正态分布

### randn

```python
df1=pd.DataFrame(npr.randn(1000))  #一维
df1[:3]
```

```python
df1.plot.line()
```

```python
df2=pd.DataFrame(npr.randn(1000,4))  #二维
df2[:3]
df2.plot.line()
```

### standard_normal

```python
df3=pd.DataFrame(npr.standard_normal(size=(1000,4)))  #size为随机抽样的次数
df3[:3]
df3.plot.line()
```

## normal正态分布
loc=期望,scale=方差，size为随机抽样的次数

```python
df4=pd.DataFrame(npr.normal(loc=1,scale=4,size=(1000,4)))  
df4[:3]
df4.plot.line()
```

## rand-标准均匀分布

```python
df5=pd.DataFrame(npr.normal(size=(1000,1)))  
df5[:3]
df5.plot.line()
```

## randint-随机整数
下限值，上限值，size随机抽样的次数

```python
df6=pd.DataFrame(npr.randint(1,1000,(1000,4)))  
df6[:3]
df6.plot.line()
```

## 随机种子-seed（）
作用是使每次执行时结果不变

```python
s1=npr.randint(1,10,5)  #重新执行时，结果会发生变化
s1[:20]
```

```python
s1+s1
```

```python
npr.seed(1000)   #添加seed（）以后，重新执行时，结果不会发生变化
s2=npr.randint(1,10,5)
s2
```
