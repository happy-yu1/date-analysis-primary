## pandas中的横向合并---merge

```python
import pandas as pd
import numpy as np
t=np.arange(12).reshape(3,4)
x1=np.hsplit(t,2)[0]
x2=np.hsplit(t,2)[1]
x1
x2
```

### 行索引相同时
#### 有相同列索引名‘code’，且该对应内容相同时

```python
y1=pd.DataFrame(x1,columns=['code','age'])
y1
```

```python
y2=pd.DataFrame(x2,columns=['kk','score'])
y2.insert(1,'code',[0,4,8])
y2
```

```python
y1.merge(y2)
```

#### 若‘code’索引下的只有数据数量不同时，会以少的为依据


```python
y1
y3=y2[:2]
y3
```

```python
pd.merge(y1,y3)
```

#### how参数可以填充
若‘code’索引下的数据数量不同时，how='outer'会自动填充，默认how=‘inner’即不显示，所以才会出现上面的情况

```python
pd.merge(y1,y3,how='outer')
```

####  on参数--- ‘code’相同，‘name’不同时
注意：‘code’的对应位置的内容相同时，会直接原地合并，若不同，on=‘code’会进行向下填充
如：y1['code']=[1,2,3],,,,y2['code']=[2,3,4]  合并后y['code]=[1,2,3,2,3,4]
若：y1['code']=[1,2,3],,,,y2['code']=[32,2,3] 合并后y['code]=[1,2,3,32]
若：y1['code']=[1,2,3],,,,y2['code']=[1,2,3]  合并后y['code]=[1,2,3

```python
y1=pd.DataFrame([[1,2,3],[3,4,9],[5,6,33]],columns=['code','hh','name'])
y2=pd.DataFrame([[1,2,0],[3,4,4],[5,7,8]],columns=['code','name','score'])
y1
y2
```

```python
y1.merge(y2,on='code',how='outer')
```

#### suffixes参数
可以变name_x，name_y为name_左边，name_右边

```python
y1.merge(y2,on='code',how='outer',suffixes=['_左边','_右边'])
```

#### code索引下的内容不同时


```python
y1=pd.DataFrame([[11,2,3],[3,4,9],[5,6,33]],columns=['code','hh','name'])
y2=pd.DataFrame([[1,2,0],[3,4,4],[5,7,8]],columns=['code','name','score'])
y1
y2
```

```python
pd.merge(y1,y2） #报错
```

操作一，，传入left_on，right_on参数

```python
# y1.merge(y2,left_on='hh',right_on='score',how='outer')---结果一样

pd.merge(y1,y2,left_on='hh',right_on='score',how='outer')
```

操作二---传入on参数

```python
y1.merge(y2,on=['code','name'],how='outer')
```

```python
y1.merge(y2,on='code',how='outer')
```

```python
y1.merge(y2,on='name',how='outer')
```

### 行索引不同时

#### 行索引数量不同
如y1的是0，1，2    y2的行索引是0,1,2,3,4

情形一，，有多个列索引名相同时

```python
y1=pd.DataFrame([[1,2,3],[3,4,9]],columns=['code','hh','name'])
y2=pd.DataFrame([[1,2,0],[3,4,4],[5,7,8],[5,7,8]],columns=['code','name','score'])
y1
y2
```

```python
y1.merge(y2,on='code',how='outer')
```

情形二--只有一个列索引名相同时---不需要添加on参数，会自动以‘code’为依据进行合并

```python
y1=pd.DataFrame([[1,2,3],[3,4,9]],columns=['code','hh','age'])
y2=pd.DataFrame([[1,2,0],[3,4,4],[5,7,8],[5,7,8]],columns=['code','name','score'])
y1
y2
```

```python
y1.merge(y2,how='outer')
```

#### 行索引名不相同时，但数量相同
#### 需要添加

```python
y1=pd.DataFrame([[1,2,3],[3,4,9],[5,7,8]],index=['a','b','c'],columns=['kkkk','hh','age'])
y2=pd.DataFrame([[1,2,0],[3,4,4],[5,7,8]],columns=['code','name','score'])
y1
y2
```

```python
y1.merge(y2,left_index=True,right_index=True,how='outer')
```

## 横向、纵向合并concat（）
以列表的形式将待合并的参数传进去，和merge不同
通过axis来判断是横向还是纵向
axis=0纵向拼接（常用）
axis=1横向拼接

### axis=0纵向拼接

```python
y1=pd.DataFrame([[1,2,3],[3,4,9],[5,7,8]],columns=['kkkk','hh','age'])
y2=pd.DataFrame([[1,2,0],[3,4,4],[5,7,8]],columns=['code','name','score'])
y1
y2
```

```python
pd.concat([y1,y2],sort=True)
```

### ignore_index参数（法一）
将以前的列索引删除,进行重新索引

```python
pd.concat([y1,y2],sort=True,ignore_index=True)
```

### 使用外部reset_index（）函数（法二）
添加drop=True参数，将以前的索引删除掉，并进行重新索引

```python
pd.concat([y1,y2],sort=True).reset_index(drop=True)
```

### 删除重复行--drop_duplicates()
适用于列索引名相同时的情况

```python
y1=pd.DataFrame([[1,2,3],[3,4,9],[5,7,8]],index=['a','b','c'],columns=['kkkk','hh','age'])
y2=pd.DataFrame([[1,2,0],[3,4,9],[5,7,8]],columns=['kkkk','hh','age'])
y1
y2
```

```python
pd.concat([y1,y2],ignore_index=True,sort=True)
```

```python
pd.concat([y1,y2],ignore_index=True,sort=True).drop_duplicates()
```
