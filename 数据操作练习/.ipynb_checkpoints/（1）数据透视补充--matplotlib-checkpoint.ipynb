**知识点扩充：**
plt的Rectangle参数：

　　　　第一个参数是坐标（x，y），即矩形的画图的起点坐标，这个起点坐标不是一味地从左下角开始画，而是对应整个图中坐标原点，即（0，0）。

　　　　第二个参数是矩形宽度

　　　　第三个坐标是矩形高度
   

```python
fig = plt.figure()
ax = fig.add_subplot(111)

rect = plt.Rectangle((0.2,0.3),0.5,0.3)
ax.add_patch(rect)
```

```python
plt.figure(figsize=(10,3))
ax1=plt.subplot(2,2,1)
ax2=plt.subplot(2,2,4)
plt.tight_layout(rect=(0.4,0.5,2,2))
# rect参数 : tuple (left, bottom, right, top),默认（0,0,1,1）
```

##### 参数ha和va设置对齐方式
ha  (horizontalalignment--水平对齐方式）='center',
va  (verticalalignment--垂直对齐方式）='center', 
transform=ax.transAxes

```python
#用zip的方法迭代
x=range(7)
y=[-1.1,2.58,-3,4,5.1,-2,3.3]
ax=plt.subplot(1,1,1)
ax.plot(x,y)

for i,j in zip(x,y):
    plt.text(i,j+0.05,'%.2f' % j,ha='center',va='top')
```

```python
#用enumerate的方法迭代(可以)
x=['a','b','c','d','e','f','g']
y=[-1.1,2.58,-3,4,5.1,-2,3.3]
ax=plt.subplot(1,1,1)
ax.plot(x,y)

for i,j in enumerate(y):
    plt.text(i,j+0.05,'%.2f' % j,ha='center',va='top')
```

##### bbox参数设置颜色和透明度
bbox=dict(facecolor='red', alpha=0.5) 以字典的方式传入

```python
x=range(7)
y=[-1.1,2.58,-3,4,5.1,-2,3.3]
ax=plt.subplot(1,1,1)
ax.plot(x,y)

for i,j in zip(x,y):
    plt.text(i,j+0.05,'%.2f' % j,ha='center',va='top',bbox={'facecolor':'red', 'alpha':0.5})
```

#### 折线图添加折点坐标--annotatie（）
当x不是从0开始的连续递增的整数时，只能用zip进行迭代，不可以用enumerate进行迭代

```python
x=[2,4,6,8,10,12,14]
y=[-1.1,2.58,-3,4,5.1,-2,3.3]
ax=plt.subplot(1,1,1)
ax.plot(x,y)

for i,j in zip(x,y):
    plt.annotate('(%s,%s)' % (i,j),xy=(i,j),ha='center',va='bottom',bbox={'facecolor':'red', 'alpha':0.5})
# for i,j in enumerate(y):
#     plt.annotate('(%s,%s)' % (i,j),xy=(i,j),ha='center',va='bottom',bbox={'facecolor':'red', 'alpha':0.5})      
```

#### bar条形图堆叠--传入bottom参数

##### 情形一：y1,y2,y3为列表
传入bottom参数时，若y3以y1,y2为底，因为y1,y2是列表形式，直接bottom=y1+y2是不可以的，相加后的元素个数和y3不匹配
需要将其转化为数组的形式，在进行相加  bottom=(np.array(y1)+np.array(y2))

```python
x=range(7)
y1=[11,58,30,40,51,63,33]
y2=[18,99,31,42,21,68,48]
y3=[7,44,38,22,61,23,68]
plt.bar(x,y1,color='b',label='第一条线')
plt.bar(x,y2,color='g',label='第二条线',bottom=y1)  #以y1未底
plt.bar(x,y3,color='r',label='第三条线',bottom=(np.array(y1)+np.array(y2))) 
plt.legend()
plt.grid()
```

##### 情形二：y1,y2,y3为一维数组
直接bottom=y1+y2即可

```python
x=range(7)
y1=np.array([11,58,30,40,51,63,33])
y2=np.array([18,99,31,42,21,68,48])
y3=np.array([7,44,38,22,61,23,68])
plt.bar(x,y1,color='b',label='第一条线')
plt.bar(x,y2,color='g',label='第二条线',bottom=y1)  #以y1未底
plt.bar(x,y3,color='r',label='第三条线',bottom=y1+y2)
plt.legend()
```

##### 情形三：不用bottom参数

```python
x=range(7)
y1=np.array([11,58,30,40,51,63,33])
y2=np.array([18,99,31,42,21,68,48])
y3=np.array([7,44,38,22,61,23,68])
plt.bar(x,y1,color='b',label='第一条线',alpha=0.6)
plt.bar(x,y2,color='g',label='第二条线',alpha=0.6)  #以y1未底
plt.bar(x,y3,color='r',label='第三条线',alpha=0.6)
plt.legend()
```
