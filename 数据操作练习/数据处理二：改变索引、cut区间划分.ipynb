```python
import pandas as pd
df=pd.read_csv('000001.SZ.csv',index_col=0)
df.head()
```

### 改变类型和索引

```python
df['trade_date'].dtype
```

```python
df['trade_date']=df['trade_date'].astype('datetime64')
df['trade_date'].head(3)
```

```python
df.set_index('trade_date',inplace=True)
df.head()
```

### 区间划分---cut  自己设置具体的划分区间
bins=[change_min,0,0.1,0.25,change_max] 

```python
df.head()
```

```python
change_min=df['change'].min()
change_max=df['change'].max()
change_min
change_max
```

```python
bins=[change_min,0,0.1,0.25,change_max]  # 划分四个区间
```

```python
c_change=pd.cut(df['change'],bins,labels=['低变化','中变化','高变化','虚假变化'])
c_change
```

```python
c_change.value_counts()  
```

### 区间划分---qcut   按中位数进行划分
 [(-0.481, -0.06] < (-0.06, -0.01] < (-0.01, 0.06] < (0.06, 0.37]]

```python
df['change'].head()
```

```python
p_change=pd.qcut(df['change'],4)
p_change.head()
```

```python
p_change.value_counts()  #计数统计
```
