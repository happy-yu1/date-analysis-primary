### 异常值处理
将‘a’列＞=60的数据和‘c’列>=100的数据作为异常值，进行剔除

```python
df1
```

#### 对数组进行描述
观察最大值，最小值，中位数，，等等或者看散点图，判断是否有异常值

```python
df1.describe()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20, 8), dpi=80)
plt.scatter(df1['a'],df1['c'])
plt.show()
```

#### 将异常值替换成中位数，或者平均数

```python
line=len(df1.index)  #  将异常值替换成均值
column=['a','c']
for i in range(line):
    for j in column:
        if df1[j][i]>=60:
            df1[j][i]=df1[j].mean()
df1
```

```python
line=len(df1.index)  #  将异常值替换成中位数
column=['a','c']
for i in range(line):
    for j in column:
        if df1[j][i]>=60:
            df1[j][i]=df1[j].quantile(0.5)
df1
```
