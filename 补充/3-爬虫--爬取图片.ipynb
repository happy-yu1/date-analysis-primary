# 百度贴吧爬取照片--用正则表达
爬取该https://tieba.baidu.com/p/2256306796上的图片

基本步骤：发起请求，获取响应内容，解析内容，保存数据

```python
import requests
from bs4 import BeautifulSoup
import re
```

## 发送请求+获取html内容
根据url发送请求，并获得相应的html内容

```python
def get_html(url):
    header={'User-Agent':'Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.70 Safari/537.36'}
    resp=requests.get(url,headers=header)
    return resp.text   #此处用的是text
```

```python
url="https://tieba.baidu.com/p/2256306796"
html=get_html(url)
html
```

## 提取图片对应的url

利用re模块，从html中获取图片对应的url，url格式如下：src中保存的url即目标url

<img class="BDE_Image" src="https://imgsa.baidu.com/forum/w%3D580/sign=43e292947c1ed21b79c92eed9d6fddae/6bfab2fb43166d228b3c16f2472309f79052d20a.jpg" width="560" height="314" pic_type="0">

```python
def get_url(html):
    jpgReg=re.compile(r'<img.+?src="(.+?\.jpg)" width')     #提取url信息的正则表达式
    # jpgReg=re.compile(r'<img.*?src="(.*?\.jpg)" width')    #结果一样
    jpgs=re.findall(jpgReg,html)   #根据c，从html中提取所有url信息
    return jpgs 
    
```

## 正则表达式说明
.表示匹配任意一个字符（除\n和\r)   +表示贪婪匹配（>=1次）   ？表示非贪婪匹配（0或1次）  *表示可有可无（0或n次），

.+? 和 .*? 匹配的结果是一样的，加上？表明非贪婪模式，匹配到即停止，不加？会一直匹配下去，一般会加上？进行非贪婪限制

```python
url="https://tieba.baidu.com/p/2256306796"
html=get_html(url)
result=get_url(html)
len(result) #  共51张图片的url
result[:5]
```

## 不加？时
只匹配到23条信息

```python
def get_url1(html):
    jpgReg=re.compile(r'<img.*src="(.*\.jpg)" width')     #提取url信息的正则表达式
    jpgs=re.findall(jpgReg,html)   #根据c，从html中提取所有url信息
    return jpgs 
url="https://tieba.baidu.com/p/2256306796"
html=get_html(url)
result=get_url1(html)
len(result)  #只匹配到23条？
result[:5]
```

## contextlib的用法

任何对象，只要正确实现了上下文管理，就可以用于with语句。实现上下文管理是通过__enter__和__exit__这两个方法实现的。

```python
class Query(object):

    def __init__(self, name):
        self.name = name

    def __enter__(self):
        print('Begin')
        return self

    def __exit__(self, exc_type, exc_value, traceback):
        if exc_type:
            print('Error')
        else:
            print('End')

    def query(self):
        print('Query info about %s...' % self.name)

with Query('Bob') as q:
    q.query()
```

如果一个对象没有实现上下文，我们就不能把它用于with语句。这个时候，可以用contextlib中的closing()来把该对象变为上下文对象

```python
from contextlib import closing #导入自动关闭请求和响应的模块
    with closing(requests.get(url,headers=header,stream = True)) as resp:
        with open (filename,'wb') as f:
            for chunk in resp.iter_content(128):
                f.write(chunk)
```

```python
def get_html(url):
    header={'User-Agent':'Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.70 Safari/537.36'}
    resp=requests.get(url,headers=header)
    return resp.text

def get_url(html):
    jpgReg=re.compile(r'<img.+?src="(.+?\.jpg)" width')     #提取url信息的正则表达式
    # jpgReg=re.compile(r'<img.*?src="(.*?\.jpg)" width')    #结果一样
    jpgs=re.findall(jpgReg,html)   #根据c，从html中提取所有url信息
    return jpgs 
    
    
def downloadJpg(imgurl,filename):
    from contextlib import closing #导入自动关闭请求和响应的模块
    with closing(requests.get(imgurl,headers=header,stream = True)) as resp:
        with open (filename,'wb') as f:
            for chunk in resp.iter_content(128):
                f.write(chunk)

def downloadJpgs(imgurls,path):  #path为文件要保存的路径
    count=1
    for url in imgurls:
        downloadJpg(url,''.join([path,'{}.jpg'.format(count)])) #
        print('下载完成第{}张图片'.format(count))
        count+=1
def downloads(url,path):
    html=get_html(url)
    urls=get_url(html)
    downloadJpgs(urls,path)
```
