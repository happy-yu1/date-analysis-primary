### 做出y1、y2关于x的图，进行对比

```python
from  matplotlib import pyplot as plt
import  random
x=range(11,20)
y1=[1,0,3,4,5,2,9,10,3]
y2=[2,1,2,0,3,5,4,4,7]
```

```python
plt.figure(figsize=(20,8),dpi=80)   #figsize尺寸，dpi像素大小
plt.figure(figsize=(20,8),dpi=90)
plt.plot(x,y1)
plt.plot(x,y2)

plt.show()
```

#### 1、添加图例和说明，以便区分哪个是y1,y2

```python
plt.figure(figsize=(20,8),dpi=80)   #figsize尺寸，dpi像素大小
plt.figure(figsize=(20,8),dpi=90)
plt.plot(x,y1,label='y1')
plt.plot(x,y2,label='y2')
plt.show()
```

#### 2、仅添加label后，图上并没有显示出图例，需要调用legend（）函数

```python
plt.figure(figsize=(20,8),dpi=80)   #figsize尺寸，dpi像素大小
plt.plot(x,y1,label='我的y1')  #添加label参数
plt.plot(x,y2,label='你的y2')
plt.legend()   #图例中的中文不显示
plt.show()
```

```python
from matplotlib import font_manager  #font_manager用来管理字体样式，
my_font=font_manager.FontProperties(fname='C:/Windows/Fonts/simsun.ttc')  #添加中文字体
plt.figure(figsize=(20,8),dpi=90)
plt.plot(x,y1,label='我的y1')
plt.plot(x,y2,label='你的y2')
plt.legend(prop=my_font)   #图例中的中文显示出来了
plt.show()
```

#### 3、通过设置 loc参数，改变图例的位置，具体情况可以  print(help(plt.legend)) 查看
        'best'          0
        'upper right'     1
        'upper left'      2
        'lower left'      3
        'lower right'     4
        ……

```python
from matplotlib import font_manager  #font_manager用来管理字体样式，
my_font=font_manager.FontProperties(fname='C:/Windows/Fonts/simsun.ttc')  #添加中文字体
plt.figure(figsize=(20,8),dpi=90)
plt.plot(x,y1,label='我的y1')
plt.plot(x,y2,label='你的y2')
plt.legend(prop=my_font,loc=2)   #图例中的中文在左上方显示出来了
plt.show()
```

```python
from matplotlib import font_manager  #font_manager用来管理字体样式，
my_font=font_manager.FontProperties(fname='C:/Windows/Fonts/simsun.ttc')  #添加中文字体
plt.figure(figsize=(20,8),dpi=90)
plt.plot(x,y1,label='我的y1')
plt.plot(x,y2,label='你的y2')
plt.legend(prop=my_font,loc=2,fontsize='small')   #图例中的中文在左上方显示出来了
plt.show()
```

#### 4、设置plt.plot（）中的参数
#### alpha 设置透明度，1表示完全不透明, linestyle设置线条风格， colo设置线条颜色， linewidth设置线条粗细

```python
from matplotlib import font_manager  #font_manager用来管理字体样式，
my_font=font_manager.FontProperties(fname='C:/Windows/Fonts/simsun.ttc')  #添加中文字体
plt.figure(figsize=(20,8),dpi=90)
plt.plot(x,y1,label='我的y1',linestyle='--',color='r')
plt.plot(x,y2,label='你的y2',linestyle='-.',color='c')
plt.legend(prop=my_font,loc=2)   #图例中的中文在左上方显示出来了
plt.show()
```
