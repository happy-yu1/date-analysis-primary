### 美国人口普查结果发现124百万人在离家相对较远的地方工作，对其上班时间进行抽样统计，得到相关结果：
#### 其中interval为上班时长，width为组距（组距不是固定的，因为上班时间在45分钟往后的人很少，仍以5为组距进行划分不合理，根据人数
#### 改变对应的组距），quantity为不同时间段统计出来的人数，如0~5分钟有836人，5~10分钟2737人……，quantity并不是原始数据，而是
#### 统计后得到的数据，若想利用如下数据得到直方图，不可以用plt.hist（）方法，而是借用条形图的方式进行转化

```python
from  matplotlib import pyplot as plt
from matplotlib import font_manager

interval = [0,5,10,15,20,25,30,35,40,45,60,90]
width = [5,5,5,5,5,5,5,5,5,15,30,60]
quantity = [836,2737,3723,3926,3596,1438,3273,642,824,613,215,47]
```

#### 1、quantity为y轴，确定x轴--做草图

```python
plt.figure(figsize=(20,8),dpi=90)
plt.bar(range(len(quantity)),quantity)
```

#### 1.2 变成直方图的形式，调整宽度。默认的是0.8，调成1试一下

```python
plt.figure(figsize=(20,8),dpi=90)
plt.bar(range(len(quantity)),quantity,width=1)
```

#### 2、修改x轴的刻度，使其变成  interval对应的刻度
#### 2.1 plt.xticks()中可以传入一个数组，只要是可以遍历的的就行


```python
plt.figure(figsize=(20,8),dpi=90)
plt.bar(range(len(quantity)),quantity,width=1)

plt.xticks(range(len(quantity)),interval)  #用interval中的数字，替换range(len(quantity)
plt.grid()
plt.show()
```

#### 2.2 x轴刻度对应的数字存在两个问题
#### （1）：0,5,10……应该往左边移动0.5个刻度
#### （2）：上班时间最长统计到150，而此处最多90，需要添加上

```python
plt.figure(figsize=(20,8),dpi=90)

x_tick=[(i-0.5) for i in range(len(quantity))]  #以便左移
plt.bar(range(len(quantity)),quantity,width=1)
plt.xticks(x_tick,interval)
plt.show()
```

```python
plt.figure(figsize=(20,8),dpi=90)

x_tick=[(i-0.5) for i in range(len(quantity)+1)]  #x_tick也添加了一个数，在ticks（）替换中保持个数一致
plt.bar(range(len(quantity)),quantity,width=1)
plt.xticks(x_tick,interval+[150])  #用添加150这个数据

plt.grid()
plt.show()
```
