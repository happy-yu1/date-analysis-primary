```python
from  matplotlib import pyplot as plt
import  random
x=range(120)
y=[random.randint(20,35) for i in range(120)]
```

#### 1、添加figure（）前后图像对比

```python
plt.plot(x,y)
plt.show()
```

```python
plt.figure(figsize=(20,8),dpi=80)   #figsize尺寸，dpi像素大小
plt.plot(x,y)
plt.show()
```

#### 2、保存图片

```python
plt.savefig('tempreture.png')
```

#### 3、修改x轴的刻度---plt.xticks（）

```python
xtick_labels = ['10点{}分'.format(i) for i in range(60)]
xtick_labels += ['11点{}分'.format(i) for i in range(60)]
plt.figure(figsize=(20,8),dpi=80)
plt.xticks(list(x),xtick_labels)  #将x轴的刻度变成字符串形式的,此方法虽然改变了刻度但是重叠了，看不清，可取个步长
plt.plot(x,y)
```

```python
plt.figure(figsize=(20,8),dpi=80)
plt.xticks(list(x)[::3],xtick_labels[::3],rotation=45)  #以3为步长,数值型和字符串格式的要同步，可将字符串进行进行45度倾斜，rotation逆时针
plt.plot(x,y)
plt.show()
```

#### 4、将未显示出来的中文显示出来

```python
from matplotlib import font_manager  #font_manager用来管理字体样式，
my_font=font_manager.FontProperties(fname='C:/Windows/Fonts/simsun.ttc')  #添加中文字体
plt.figure(figsize=(20,8),dpi=90)
plt.xticks(list(x)[::4],xtick_labels[::4],rotation=45,fontProperties=my_font)
plt.plot(x,y)
plt.show()   #中文字体显示出来了
```

#### 5、添加x、y轴和标题信息

```python
from matplotlib import font_manager  #font_manager用来管理字体样式，
my_font=font_manager.FontProperties(fname='C:/Windows/Fonts/simsun.ttc')  #添加中文字体
plt.figure(figsize=(20,8),dpi=90)
plt.xticks(list(x)[::4],xtick_labels[::4],rotation=45,fontProperties=my_font)


plt.xlabel('时间')
plt.ylabel('气温 单位（摄氏度）')
plt.title('10点到12点每分钟气温变化情况')
plt.plot(x,y)
plt.show()
```

#### 6、添加fontProperties参数，将详细信息中的中文显示出来

```python
from matplotlib import font_manager  #font_manager用来管理字体样式，
my_font=font_manager.FontProperties(fname='C:/Windows/Fonts/simsun.ttc')  #添加中文字体
plt.figure(figsize=(20,8),dpi=90)
plt.xticks(list(x)[::4],xtick_labels[::4],rotation=45,fontProperties=my_font)

plt.xlabel('时间',fontProperties=my_font)
plt.ylabel('气温 单位（摄氏度）',fontProperties=my_font)
plt.title('10点到12点每分钟气温变化情况',fontProperties=my_font)
plt.plot(x,y)
plt.show()
```

#### 7、添加网格--grid（）函数、设置透明度（设置alpha参数）

```python
from matplotlib import font_manager  #font_manager用来管理字体样式，
my_font=font_manager.FontProperties(fname='C:/Windows/Fonts/simsun.ttc')  #添加中文字体
plt.figure(figsize=(20,8),dpi=90)
plt.xticks(list(x)[::4],xtick_labels[::4],rotation=45,fontProperties=my_font)

plt.xlabel('时间',fontProperties=my_font)
plt.ylabel('气温 单位（摄氏度）',fontProperties=my_font)
plt.title('10点到12点每分钟气温变化情况',fontProperties=my_font)
plt.plot(x,y)
plt.grid()  #添加网格
plt.show()
```

```python
from matplotlib import font_manager  #font_manager用来管理字体样式，
my_font=font_manager.FontProperties(fname='C:/Windows/Fonts/simsun.ttc')  #添加中文字体
plt.figure(figsize=(20,8),dpi=90)
plt.xticks(list(x)[::4],xtick_labels[::4],rotation=45,fontProperties=my_font)

plt.xlabel('时间',fontProperties=my_font)
plt.ylabel('气温 单位（摄氏度）',fontProperties=my_font)
plt.title('10点到12点每分钟气温变化情况',fontProperties=my_font)
plt.plot(x,y)
plt.grid(alpha=1,linestyle='-.',color='r',linewidth=5)   
# alpha 设置透明度，1表示完全不透明,linestyle设置线条风格，colo设置线条颜色，linewidth设置线条粗细
plt.show()    
```
