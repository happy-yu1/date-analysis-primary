#### 1、散点图x  y不相干

```python
from  matplotlib import pyplot as plt
import numpy as np
from matplotlib import font_manager

N=1000
x=np.random.randn(N)   #随机生成1000个标准正太分布的数据
y=np.random.randn(N)

plt.figure(figsize=(20,8),dpi=90)
plt.scatter(x,y,c='r',alpha=0.4,s=600,marker='v')
#通过调整alpha和s的大小，可以观察到在哪些地方重叠的较多，因为重叠部分颜色会加深
```

```python
import numpy as np
import matplotlib
import matplotlib.dates as mdates
from matplotlib import font_manager
import csv,random
from  matplotlib import pyplot as plt
```

```python
header=['date','temper']
x = ['10/{}/2015'.format(i) for i in range(1,32)]+['11/{}/2015'.format(i) for i in range(1,31)]
y=[random.randint(200,350) for i in range(61)]
totle=[(x[i],y[i]) for i in range(61)]

with open('temper.csv','w',newline='') as f:
    writer=csv.writer(f)
    writer.writerow(header)
    writer.writerows(totle)
```

```python
def datestr2num(bytedates):
    return mdates.time.strptime(bytedates.decode('utf-8'),'%m/%d/%Y')

date=np.loadtxt('temper.csv',delimiter=',',converters={0:datestr2num},skiprows=1,usecols=(0))
```
