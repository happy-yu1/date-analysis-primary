#### 1、组数和组距

```python
print(max(a),min(a),max(a)-min(a))
bin_width=5  #组距
bin_nums=(max(a)-min(a))//bin_width
print(bin_nums)      #15组，组数必须为整数
```

#### 2、草图--传入a和组数

```python
plt.figure(figsize=(20,8),dpi=90)
plt.hist(a,bin_nums)
plt.show()
```

#### 3、修改x轴刻度

#### 3.1、修改x轴刻度--以最大值和最小值作为x轴刻度,并以组距为步长

```python
plt.figure(figsize=(20,8),dpi=90)
plt.hist(a,bin_nums)

plt.xticks(list(range(min(a),max(a)))[::bin_width],rotation=45) 
plt.show()
```

#### 3.2、range(78,156)并没有将最大值156包含进去，所以x轴刻度中要将max（a）+bin_width

```python
plt.figure(figsize=(20,8),dpi=90)
plt.hist(a,bin_nums)

plt.xticks(list(range(min(a),max(a)+bin_width))[::bin_width],rotation=45) 
plt.show()
```

#### 3.3、观察上图，出现刻度错位的情况，通过画网格来看一下
#### 从88就开始往右或者往左偏移了，并不是我们给的组距，原因是,max(a)-min(a)=78，所选的组距不能被78整除，只有当整除时，才不会发生错位

```python


plt.figure(figsize=(20,8),dpi=90)
plt.hist(a,bin_nums)

plt.xticks(list(range(min(a),max(a)+bin_width))[::bin_width],rotation=45) 
plt.grid(alpha=0.6,color='r')
plt.show()  
```

####    ----bin_width=6时----没有发生错位现象 

```python
bin_width=6  #组距
bin_nums=(max(a)-min(a))//bin_width

plt.figure(figsize=(20,8),dpi=90)
plt.hist(a,bin_nums)

plt.xticks(list(range(min(a),max(a)+bin_width))[::bin_width],rotation=45) 
plt.grid(alpha=0.6,color='r')
plt.show()  
```

#### 3.3、频数直方图变频率直方图，修改density参数即可，默认为0（表示为频数图）1（表示为频率图）

```python
bin_width=6  #组距
bin_nums=(max(a)-min(a))//bin_width

plt.figure(figsize=(20,8),dpi=90)
plt.hist(a,bin_nums,density=1)

plt.xticks(list(range(min(a),max(a)+bin_width))[::bin_width],rotation=45) 
plt.grid(alpha=0.6,color='r')
plt.show() 
```
