```python
class Sample(object):
    """
        样本点类
    """
    def __init__(self, coords):
        self.coords = coords    # 样本点包含的坐标
        self.n_dim = len(coords)    # 样本点维度

    # 将coords转成str输出
    # 方便查看
    def __repr__(self):
        return str(self.coords)
lower = 0
upper = 200
n_dim=2
n_samples=1000
samples = [Sample([random.uniform(lower, upper) for _ in range(n_dim)]) for _ in range(n_samples)]
```
