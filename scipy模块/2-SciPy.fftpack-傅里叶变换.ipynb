使用sciPy.fftpack模块可以快速的进行傅里叶的转换
需要先下载scipy包

s速度比传统的方法要快，图片是二维数据，要使用fftpack的二维转变方法

 登月照片降噪

```python
from  matplotlib import pyplot as plt
import numpy as np
import pandas as pd 
import scipy.fftpack as fft
```

# 读取照片
imread读取，，imshow展示图片（热图），imsave保存图片

```python
moon_data=plt.imread("moon.png")  #返回的是numpy.ndarray类型的数据
moon_data[:2]
```

```python
plt.figure(figsize=(15,9))
plt.imshow(moon_data)  #展示照片（热图？）
```

fft下有fft和fft2等多种变换方法，fft是针对一维数组，fft2针对多维数组，因为moon_data为二维数组，所以要用fft2进行处理

```python
moon_data_fft=fft.fft2(moon_data)
moon_data_fft
```

```python
moon_data_fft_2=np.where(np.abs(moon_data_fft)>3e3,0,moon_data_fft)
```

```python
moon_data_ifft=fft.ifft2(moon_data_fft_2)
```

```python
moon_data_real=np.real(moon_data_ifft)
```

```python
plt.figure(figsize=(15,9))
plt.imshow(moon_data_real)
```
