## 函数形式的调用方法
后面调用积分函数时，y不可以，但是f可以，所以用函数的形式

```python
x = np.linspace(-1, 1, 1000)
f= lambda x: (1-x ** 2)** 0.5  
```

```python
f(0.35)
```

```python
plt.plot(x,f(x))
```

## 求积分--integrate
调用里面的quad方法，会返回两个值，第一个值为得到的结果，后面得到的是一个估计值

```python
 import scipy.integrate as integrate
```

```python
integrate.quad(f,-1,1)   #计算-1,1之间的积分，
```

```python
s1,est=integrate.quad(f,-1,1)
s1
```

```python
pi=2*s1
pi
```
