#### 5.2 fillna（）填充nan

```python
b={'name':['z3','li4','w5'],'age':[23,16,27],'socre':[78,81,48]}
t=pd.DataFrame(b)
t
```

```python
t.iloc[1,1]=np.nan
t.iloc[2,2]=np.nan
t
```

```python
t.fillna(100)
```

```python
t.mean()
```

```python
t.fillna(t.mean())  #将所有的nan均替换成了对应的均值
```

```python
t['age'].fillna(t['age'].mean())  #只将age列中的NaN提换成均值
```

```python
t['age']=t['age'].fillna(t['age'].mean())  #重新复制
t
```
